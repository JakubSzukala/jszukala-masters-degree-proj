```python
import numpy as np
```

```python
domains_names, value_counts = np.unique(train_df_images['domain'], return_counts=True)
print(domains_names)
value_counts
```

```python
weights = 1 / value_counts
weights
```

```python
train_df_images['weight'] = train_df_images['domain'].apply(lambda x: weights[np.where(domains_names == x)[0][0]])
train_df_images
```

```python
from torch.utils.data import WeightedRandomSampler
from torch.utils.data import DataLoader
```

```python
from data.adapter import GwhdToYoloAdapter
from yolov7.dataset import Yolov7Dataset
from yolov7.dataset import yolov7_collate_fn
train_adapter = GwhdToYoloAdapter(os.path.join(DATASET_ROOT_DIR, 'images'), train_df_annotations, None)
yolo_train_ds = Yolov7Dataset(train_adapter)
```

```python
print(len(train_df_images))
print(len(yolo_train_ds))
sampler = WeightedRandomSampler(
    weights=train_df_images['weight'],
    num_samples=len(train_df_images['weight']),
    replacement=True
)
dl = DataLoader(yolo_train_ds, batch_size=1, sampler=sampler)
```

```python
from tqdm import tqdm
def print_dataloader(dl, n_batches=3):
    domain_counts = {
        'Arvalis_1' : 0,
        'Arvalis_2' : 0,
        'Arvalis_3' : 0,
        'Arvalis_4' : 0,
        'Arvalis_5' : 0,
        'Arvalis_6' : 0,
        'Arvalis_7' : 0,
        'Arvalis_8' : 0,
        'Arvalis_9' : 0,
        'Arvalis_10' : 0,
        'Arvalis_11' : 0,
        'Arvalis_12' : 0,
        'ETHZ_1' : 0,
        'Inrae_1' : 0,
        'NMBU_1' : 0,
        'NMBU_2' : 0,
        'Rres_1' : 0,
        'ULiège-GxABT_1' : 0
    }
    counter = 0
    indices_seen = []
    for batch in tqdm(dl):
        images, labels, indices, sizes = batch
        indices_seen.extend([int(idx) for idx in indices])
        for idx in indices:
            domain_counts[train_df_images.iloc[int(idx)]['domain']] += 1
        #{k : 0 for k in domain_counts.keys()}
        counter += 1
        if counter == n_batches:
            return indices_seen, domain_counts
        #idxs = batch[0][:, 0].tolist()
        #classes = batch[0][:, 1]
        #class_ids, class_counts = classes.unique(return_counts=True)
        #class_ids = set(class_ids.tolist())
        #class_counts = class_counts.tolist()
        #print(f"Batch {i}: ({class_ids})")
```

```python
dl_upsampled = DataLoader(
    yolo_train_ds,
    batch_size=32,
    collate_fn=yolov7_collate_fn,
    sampler=sampler
)
#visualise_dataloader(dl_original)
indices_seen_upsampled, domain_counts_upsampled = print_dataloader(dl_upsampled, n_batches=int(len(train_df_images) // 32))
```

```python
from matplotlib import pyplot as plt
plt.bar(domain_counts_upsampled.keys(), domain_counts_upsampled.values())
plt.xticks(rotation=90)
print(len(indices_seen_upsampled))
print(np.sum(list(domain_counts_upsampled.values())))
```

```python
dl_original = DataLoader(
    yolo_train_ds,
    batch_size=32,
    collate_fn=yolov7_collate_fn,
    sampler=None
)
indices_seen_original, domain_counts_original = print_dataloader(dl_original, n_batches=int(len(train_df_images) // 32))
```

```python
plt.bar(domain_counts_original.keys(), domain_counts_original.values())
plt.xticks(rotation=90)
print(len(indices_seen_original))
print(np.sum(list(domain_counts_original.values())))
```

```python
from collections import Counter

image_counts_df = (
    pd.merge(
        train_df_images.reset_index(drop=True)
        .reset_index()
        .rename(columns={"index": "image_idx"}),
        pd.DataFrame.from_records(
            {"image_idx": k, "seen_count": v} for k, v in Counter(indices_seen_upsampled).items()
        ),
        how="left",
    )
    .fillna(0)
    .sort_values("seen_count", ascending=False)
    .reset_index(drop=True)
)
image_counts_df
```

```python
import seaborn as sns

fig, ax = plt.subplots(figsize=(10, 10))
for domain in domains_names:
    sns.ecdfplot(image_counts_df.query(f'domain == "{domain}"').seen_count, ax=ax, label=domain)
ax.set_xlabel("Number of times seen during training", fontsize=12)
ax.set_ylabel("Proportion of images in the dataset", fontsize=12)
ax.legend(fontsize=12)
ax.set_title("ECDF Plot showing how often images were seen during training")
fig.tight_layout()
fig.show()
```
