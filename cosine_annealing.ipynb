```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Define parameters for the scheduler
max_epochs = 100
warmup_epochs = 10
eta_min = 0.001
eta_max = 0.01
k_decays = np.linspace(0.0, 3.0, 5)[1:]

# Calculate learning rates for each epoch
for k_decay in k_decays:
    learning_rates = []
    for epoch in range(max_epochs):
        if epoch < warmup_epochs:
            lr = ((epoch + 1) / warmup_epochs) * eta_max
        else:
            lr = eta_min + (eta_max - eta_min) * (1 + math.cos(math.pi * (epoch - warmup_epochs)**k_decay / (max_epochs - warmup_epochs)**k_decay)) / 2
        learning_rates.append(lr)

    # Plot the learning rate schedule
    plt.plot(np.arange(max_epochs), learning_rates, label='k_decay = {}'.format(k_decay))
plt.xlabel('Epochs')
plt.ylabel('Learning Rate')
plt.title('Cosine Annealing Learning Rate Schedule')
plt.legend()
plt.show()
```
